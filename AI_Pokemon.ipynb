```python
from sklearn.cluster import BisectingKMeans, KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, precision_recall_fscore_support
from sklearn.svm import SVC
```

Machine Learning

```python
#datitos
X = pk_df[['attack', 'defense', 'hp', 'sp.atk', 'sp.def', 'speed']]
y = pk_df['type1'] 
#dividir (y conquistar) los conjuntos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
#escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# entrenar y evluar un clasificador base
clf_base = RandomForestClassifier(n_estimators=100, random_state=42) #
clf_base.fit(X_train_scaled, y_train)
y_pred_base = clf_base.predict(X_test_scaled)
print("Clasificador Base:")
print(classification_report(y_test, y_pred_base))
print(f"Accuracy: {accuracy_score(y_test, y_pred_base):.4f}")

# es la hora del clustering 
kmeans = KMeans(n_clusters=16, random_state=42)
cluster_train = kmeans.fit_predict(X_train_scaled)
cluster_test = kmeans.predict(X_test_scaled)

#Cluster como features
X_train_with_clusters = np.column_stack((X_train_scaled, cluster_train))
X_test_with_clusters = np.column_stack((X_test_scaled, cluster_test))
#entrena un nuevo clasfcador de clusters
clf_with_clusters = RandomForestClassifier(n_estimators=100, random_state=42)
clf_with_clusters.fit(X_train_with_clusters, y_train)
#evaluar(again)
y_pred_with_clusters = clf_with_clusters.predict(X_test_with_clusters)
print("\nClasificador con Clusters:")
print(classification_report(y_test, y_pred_with_clusters))
print(f"Precision: {accuracy_score(y_test, y_pred_with_clusters):.4f}")
# compara la importancia de los features
feature_names = list(X.columns) + ['cluster']
importances = clf_with_clusters.feature_importances_
for feature, importance in sorted(zip(feature_names, importances), key=lambda x: x[1], reverse=True):
    print(f"{feature}: {importance:.4f}")

## TO-DO: Agregar filtros
```

```python
stats = pk_df[['hp', 'attack', 'defense', 'speed', 'sp.atk', 'sp.def']]
n_clusters_list = [4, 8, 16]

#comparar algoritmos
clustering_algorithms = {
    "Bisecting K-Means": BisectingKMeans,
    "K-Means": KMeans,
}

fig, axs = plt.subplots(
    len(clustering_algorithms), len(n_clusters_list), figsize=(15, 8)
)
axs = axs.T

for i, (algorithm_name, Algorithm) in enumerate(clustering_algorithms.items()):
    for j, n_clusters in enumerate(n_clusters_list):
        algo = Algorithm(n_clusters=n_clusters, random_state=0, n_init=3)
        algo.fit(stats)
        
        #  PCA para reducir a 2 dimensiones
        pca = PCA(n_components=2)
        stats_2d = pca.fit_transform(stats)
        centers_2d = pca.transform(algo.cluster_centers_)
        
        axs[j, i].scatter(stats_2d[:, 0], stats_2d[:, 1], s=10, c=algo.labels_, cmap='viridis')
        axs[j, i].scatter(centers_2d[:, 0], centers_2d[:, 1], c="r", s=50, marker='x')
        axs[j, i].set_title(f"{algorithm_name}: {n_clusters} clusters")

for ax in axs.flat:
    ax.label_outer()
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
#por ataque y def
fig, axs = plt.subplots(
    len(clustering_algorithms), len(n_clusters_list), figsize=(15, 8)
)
axs = axs.T

for i, (algorithm_name, Algorithm) in enumerate(clustering_algorithms.items()):
    for j, n_clusters in enumerate(n_clusters_list):
        algo = Algorithm(n_clusters=n_clusters, random_state=0, n_init=3)
        algo.fit(stats)
        #
        axs[j, i].scatter(stats['attack'], stats['defense'], s=10, c=algo.labels_, cmap='viridis')
        axs[j, i].scatter(algo.cluster_centers_[:, 0], algo.cluster_centers_[:, 1], c="r", s=50, marker='x')
        axs[j, i].set_title(f"{algorithm_name}: {n_clusters} clusters")
        axs[j, i].set_xlabel('Attack')
        axs[j, i].set_ylabel('Defense')

plt.tight_layout()
plt.show()

# se ve horrible xD
```

Voy a hacer comparacion con distintos algoritmos de aprendizaje supervisado, usare  Random Forest (como antes), Gradient Boost y Support Vector Machine kon un kernel n/lineal, concretamente gaussiano (o radial basis funtion, rbf para los amigos)


```python
import warnings
#warnings.filterwarnings("always")

r_state=0 #sin randim state, por ahora

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=r_state)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
kmeans = KMeans(n_clusters=16, random_state=r_state)
cluster_train = kmeans.fit_predict(X_train_scaled)
cluster_test = kmeans.predict(X_test_scaled)
X_train_with_clusters = np.column_stack((X_train_scaled, cluster_train))
X_test_with_clusters = np.column_stack((X_test_scaled, cluster_test))
classifiers = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=r_state),
    'SVM': SVC(kernel='rbf', random_state=r_state),#kernel radial
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=r_state)
}

results = []
for name, clf in classifiers.items():
    print(f"\n{name}:")
   
    clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    accuracy_without = accuracy_score(y_test, y_pred)
    precision_without, recall_without, _, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted', zero_division=0)
    print("sin Clusters:")
    print(f"Exactitud: {accuracy_without:.4f}")
    print(f"Precision: {precision_without:.4f}")
    print(f"Recall: {recall_without:.4f}")
   
    clf.fit(X_train_with_clusters, y_train)
    y_pred_with_clusters = clf.predict(X_test_with_clusters)
    accuracy_with = accuracy_score(y_test, y_pred_with_clusters)
    precision_with, recall_with, _, _ = precision_recall_fscore_support(y_test, y_pred_with_clusters, average='weighted', zero_division=0)
    print("con Clusters:")
    print(f"Exactitud: {accuracy_with:.4f}")
    print(f"Precision: {precision_with:.4f}")
    print(f"Recall: {recall_with:.4f}")
    
    results.append({
        'Classifier': name,
        'Exactitud sin Clusters': accuracy_without,
        'Precision sin Clusters': precision_without,
        'Recall sin Clusters': recall_without,
        'Exactitud con Clusters': accuracy_with,
        'Precision con Clusters': precision_with,
        'Recall con Clusters': recall_with
    })

results_df = pd.DataFrame(results)

# Grfico de barras comparativo
plt.figure(figsize=(15, 8))
results_df.set_index('Classifier').plot(kind='bar', y=['Exactitud sin Clusters', 'Exactitud con Clusters', 
                                                       'Precision sin Clusters', 'Precision con Clusters'])
plt.title('Comparacin de Mtricas con y sin Clusters')
plt.ylabel('Valor')
plt.legend(['Exactitud Sin Clusters', 'Exactitud Con Clusters', 'Precision Sin Clusters', 'Precision Con Clusters'])
plt.tight_layout()
plt.show()

```

```python

# Anlisis detallado - Granadient Boosting -
best_clf = GradientBoostingClassifier(n_estimators=100, random_state=r_state)
best_clf.fit(X_train_with_clusters, y_train)
y_pred_best = best_clf.predict(X_test_with_clusters)
print("\nAnlisis detallado del mejor clasificador\n-Gradient Boosting con clusters-):")
print(classification_report(y_test, y_pred_best, zero_division=0))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 16))

feature_names = list(X.columns) + ['cluster']
importances = best_clf.feature_importances_
importances_df = pd.DataFrame({'feature': feature_names, 'importancia': importances})
importances_df = importances_df.sort_values('importancia', ascending=False)

# importancia de caracteristicas
sns.barplot(x='importancia', y='feature', data=importances_df, ax=axes[0, 0])
axes[0, 0].set_title('Importancia de las Caractersticas\n-Gradient Boosting-')
axes[0, 0].set_xlabel('Importancia')

# M Conf
cm = confusion_matrix(y_test, y_pred_best)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0, 1])
axes[0, 1].set_title('Matriz de Confusin\n-Gradient Boosting-')
axes[0, 1].set_xlabel('Prediccin')
axes[0, 1].set_ylabel('Valor real')

# comparar presicion 
gb_without_clusters = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_without_clusters.fit(X_train_scaled, y_train)
y_pred_without = gb_without_clusters.predict(X_test_scaled)

accuracies = [accuracy_score(y_test, y_pred_without), accuracy_score(y_test, y_pred_best)]
labels = ['Sin Clusters', 'Con Clusters']
axes[1, 0].bar(labels, accuracies)
axes[1, 0].set_title('Comparacin de Precision\n-Gradient Boosting-')
axes[1, 0].set_ylabel('Precision')
axes[1, 0].set_ylim(0, 1)
for i, v in enumerate(accuracies):
    axes[1, 0].text(i, v, f'{v:.4f}', ha='center', va='bottom')

# Rendimiento por tipo primario de pokemon
report = classification_report(y_test, y_pred_best, output_dict=True, zero_division=0)
df_report = pd.DataFrame(report).transpose()
df_report = df_report.iloc[:-3]  # mtricas globales (-3 filas)
df_report = df_report.sort_values('f1-score', ascending=False)

sns.barplot(x=df_report.index, y='f1-score', data=df_report, ax=axes[1, 1])
axes[1, 1].set_title('Rendimiento por tipo primario de Pokmon \n-Gradient Boosting-')
axes[1, 1].set_xticklabels(df_report.index, rotation=90)
axes[1, 1].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
# metodo de analisis y plot con funcion estandarizada, por que da pereza rehacerla para cada caso (y despues la puedo reutilizar :3)

def detailed_classifier_analysis(clf, X_train, X_test, y_train, y_test, classifier_name):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(f"\nAnlisis detallado de\n-{classifier_name}-:")
    print(classification_report(y_test, y_pred, zero_division=0))
    return clf

def plot_classifier_results(clf, X_train, X_test, y_train, y_test, classifier_name):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    feature_importances = getattr(clf, 'feature_importances_', None)
    feature_names = list(X.columns) + ['cluster']
    
    fig, axes = plt.subplots(2, 2, figsize=(20, 16))
    
    if feature_importances is not None:
        feature_imp = pd.DataFrame({'feature': feature_names, 'importance': feature_importances})
        feature_imp = feature_imp.sort_values('importance', ascending=False)
        sns.barplot(x='importance', y='feature', data=feature_imp, ax=axes[0, 0])
        axes[0, 0].set_title(f'Importancia de las Caractersticas\n-{classifier_name}-')
        axes[0, 0].set_xlabel('Importancia')
    
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0, 1])
    axes[0, 1].set_title(f'Matriz de Confusin\n-{classifier_name}-')
    axes[0, 1].set_xlabel('Prediccin')
    axes[0, 1].set_ylabel('Valor real')
    
    gb_without_clusters = clf
    gb_without_clusters.fit(X_train_scaled, y_train)
    y_pred_without = gb_without_clusters.predict(X_test_scaled)

    accuracies = [accuracy_score(y_test, y_pred_without), accuracy_score(y_test, y_pred)]
    labels = ['Sin Clusters', 'Con Clusters']
    axes[1, 0].bar(labels, accuracies)
    axes[1, 0].set_title(f'Comparacin de Precisin\n-{classifier_name}-')
    axes[1, 0].set_ylabel('Precisin')
    axes[1, 0].set_ylim(0, 1)
    for i, v in enumerate(accuracies):
        axes[1, 0].text(i, v, f'{v:.4f}', ha='center', va='bottom')

    report = classification_report(y_test, y_pred, output_dict=True, zero_division=0)
    df_report = pd.DataFrame(report).transpose()
    df_report = df_report.iloc[:-3]  # Eliminar las ltimas 3 filas (mtricas globales)
    df_report = df_report.sort_values('f1-score', ascending=False)
    
    sns.barplot(x=df_report.index, y='f1-score', data=df_report, ax=axes[1, 1])
    axes[1, 1].set_title(f'Rendimiento por tipo primario de Pokmon\n-{classifier_name}-')
    axes[1, 1].set_xticks(range(len(df_report.index)))
    axes[1, 1].set_xticklabels(df_report.index, rotation=90)
    axes[1, 1].set_ylim(0, 1)

    plt.tight_layout()
    plt.show()
```

```python
# Random Forest

rf_clf = RandomForestClassifier(n_estimators=100, random_state=r_state)
detailed_classifier_analysis(rf_clf, X_train_with_clusters, X_test_with_clusters, y_train, y_test, 'Random Forest')
plot_classifier_results(rf_clf, X_train_with_clusters, X_test_with_clusters, y_train, y_test, 'Random Forest')
```

```python
# SVM de kernel radial
svm_clf = SVC(kernel='rbf', random_state=r_state)
detailed_classifier_analysis(svm_clf, X_train_with_clusters, X_test_with_clusters, y_train, y_test, 'SVM')
plot_classifier_results(svm_clf, X_train_with_clusters, X_test_with_clusters, y_train, y_test, 'SVM')
#revisar y reparar situacion con el primer grafico, es el peor de todos, debo probar otros kernels
```
